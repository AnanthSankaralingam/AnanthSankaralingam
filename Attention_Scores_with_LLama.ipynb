```python
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_1.bin" # the model is in bin format
```

```python
from transformers import LlamaTokenizer
```

```python
from huggingface_hub import hf_hub_download
from llama_cpp import Llama
```

```python
model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

```python
# model pruned to first layer only
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2, # CPU cores
    n_batch=512, # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.
    n_gpu_layers=1 # ONLY USE FIRST ATTENTION LAYER
    )
```

```python
lcpp_llm.params.n_gpu_layers
#see layers
```

```python
prompt = "Please help! I need assistance with finding a reliable source for purchasing high-quality, affordable, and sustainable products online?!?! Can you provide me with a list of trusted websites and stores that offer eco-friendly, non-toxic, and socially responsible items?!?! Additionally, what are some tips for avoiding scams and ensuring safe delivery?!?!"

# Convert the prompt to bytes
prompt_bytes = prompt.encode('utf-8')

# Tokenize the prompt
tokenized_prompt = lcpp_llm.tokenize(text=prompt_bytes, add_bos=True)

# Print the tokenized prompt
print(tokenized_prompt)
```
